```python
df_march_20 = df[df["date"].str.match("2020-03")]
df_march_21 = df[df["date"].str.match("2021-03")]
df_march_22 = df[df["date"].str.match("2022-03")]

march_cases_20 = df_march_20["new_cases"]
march_cases_21 = df_march_21["new_cases"]
march_cases_22 = df_march_22["new_cases"]

months = [ "March 2020", "March 2021", "March 2022"]
box_data = [ march_cases_20, march_cases_21, march_cases_22 ]

# the actual data
from random import uniform as urand
eps = 0.1
x_20 = [urand(1-eps, 1+eps) for i in range(0, len(march_cases_20))]
x_21 = [urand(2-eps, 2+eps) for i in range(0, len(march_cases_21))]
x_22 = [urand(3-eps, 3+eps) for i in range(0, len(march_cases_22))]

plt.plot(x_20, march_cases_20, 'bo')
plt.plot(x_21, march_cases_21, 'ro')
plt.plot(x_22, march_cases_22, 'go')
plt.boxplot(box_data, labels = months, sym='')

plt.title("Monthly distribution of new COVID-19 cases in NYS")
plt.xlabel("Month")
plt.ylabel("New Cases")




```
